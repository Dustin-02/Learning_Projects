```python
## Libraries
# not needed
import numpy as np

import pandas as pd

# for clarification
import matplotlib.pyplot as plt
```

```python
# Read the data into a data frame.
df = pd.read_csv('pitching.csv')
df.head(5)
```

```python
# Display all pitchers in 2015 under each team sorted by their last names.
pitchers_2015 = df[df['yearID'] == 2015]
result_2 = pitchers_2015.sort_values(by = ['teamID', 'nameLast'])
result_2[['yearID', 'teamID', 'nameLast']].head(5)
```

```python
# Calculate the average ERA for each team, list the teams by the average ERA in a descending order.
ERA_avr_by_team = df.groupby('teamID')['ERA'].mean().sort_values(ascending = False)
ERA_avr_by_team.head(10)

# # Visualize
# plt.plot(ERA_avr_by_team.values)
# plt.show()
```

```python
## Calculate the total number of home runs allowed (HRs) for each team, and list the team by the count in an ascending order.
HR_count_by_team = df.groupby('teamID')['HR'].sum().sort_values()
HR_count_by_team

## For visualization
# plt.figure(figsize = (16, 6))
# plt.plot(HR_count_by_team.index, HR_count_by_team.values)
# plt.xticks(rotation=90, size = 10)
# plt.show() 
```

```python
## Calculate each team’s average ERA from 2011 to 2015,
## and list the result as a table, where each column is a year, each row is a team.

# 11 - 15 year filter
df_11to15 = df[df['yearID'].between(2011, 2015)]
# Calc ERA average
ERA_mean_by_team_11to15 = df_11to15.groupby(['teamID', 'yearID'])['ERA'].mean().reset_index()
# Change into pivot table
team_ERA_table = ERA_mean_by_team_11to15.pivot(index='teamID', columns='yearID', values='ERA')

# Show result
team_ERA_table.head(5)
```

```python
### Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015,
### and list the result as a table, where each column is a league and each row is a year.

## For AL and NL from 11 - 15 only
df_filter = df_11to15[df_11to15['lgID'].isin(['AL','NL'])]
# Test: df_filter['lgID'].unique()

## Calc mean ERA
ERA_avrg_AL_and_NL = df_filter.groupby(['lgID', 'yearID'])['ERA'].mean().reset_index()
# Test: ERA_avrg_AL_and_NL

## Convert into table 
league_ERA_pivot = ERA_avrg_AL_and_NL.pivot(index='yearID', columns='lgID', values='ERA')
league_ERA_pivot
```

```python
### List the pitchers who had the most wins in each of the years from 2011 to 2015.

# Max win by year
winner_by_year = df_11to15.groupby('yearID')['W'].max().reset_index()

# Merge on W and yearID of winner_by_year
winner_list = pd.merge(winner_by_year, df_11to15, on= ['yearID', 'W'])
winner_list

## Apply highlighting to the DataFrame
# highlight_winner_list = winner_list.style.applymap(lambda x: 'background-color: yellow', subset=['W'])
# highlight_winner_list
```
